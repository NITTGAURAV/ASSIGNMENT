QUESTION 2

Scatter plot is a graphical representation of x and y axis data in graph intersection of x and y data represented by star on graph

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("graph")
```

QUESTION 4

Bar plot use to compare different set of data among different groups as well as it represent relationship between categorical and numerical data

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```
